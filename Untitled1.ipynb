```python
img_file = os.path.join(img_dir,'E24W-1_C5_09.czi')

aicsfile = AICSImage(img_file)
print(aicsfile.scenes)
#print(aicsfile.current_scene)
# AICSImage read the mosaic image as stitched; scenes can be accessed by set_scene;
# For RGB/BGR images, "S" dimension no longer represent "Scene" but "Samples" dimension (RGB / BGR) 

aicsfile.set_scene(18)
print(aicsfile.current_scene)
print(aicsfile.dims)
#scene18, shp = img.read_image(S=18)
scene18 = aicsfile.get_image_data("YXS", T=0,C=0,Z=0) # numpy.ndarray
print(scene18.shape)
#print(scene18[:,:,0])
print(type(scene18))
scene18_tran = skimage.util.img_as_float(scene18)
print("50th percentile: ",
       np.percentile(scene18_tran.ravel(), 50))
print("25th percentile: ",
       np.percentile(scene18_tran.ravel(), 25))
print("75th percentile: ",
       np.percentile(scene18_tran.ravel(), 75))
print("90th percentile: ",
       np.percentile(scene18_tran.ravel(), 90))
print("95th percentile: ",
       np.percentile(scene18_tran.ravel(), 95))
print("98th percentile: ",
       np.percentile(scene18_tran.ravel(), 98))
#print(scene18_tran[:,:,0])
#plt.imshow(scene18)
#plt.show()
# plt.imshow(scene18_tran, cmap="gray")
imgplot = plt.imshow(scene18)
imgplot.set_clim(0, 215)
# plt.hist(scene18_tran.ravel(), bins=256, range=(0, 1))
# plt.show()
# np.max(scene18_tran.ravel())
#fig, ax = plt.subplots()
#plt.imshow(scene18)
#plt.imshow(scene18)
#plt.show()

colors = ["red", "green", "blue"]
plt.figure()
plt.xlim([0, 256])
for i in range(0,3,1):
    histogram, bin_edges = np.histogram(
        scene18[:, :, i], bins=256, range=(0, 256)
    )
    plt.plot(bin_edges[0:-1], histogram, color=colors[i])

plt.title("Color Histogram")
plt.xlabel("Color value")
plt.ylabel("Pixel count")
plt.show()
```
